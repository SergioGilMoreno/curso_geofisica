## Datos

```python
import pandas as pd
import numpy as np

# Generar datos aleatorios para el DataFrame
np.random.seed(0)  # Establecer una semilla para reproducibilidad

# Crear un DataFrame con 20 filas y 10 columnas
df = pd.DataFrame({
    'ID_producto': np.arange(1, 21),
    'Nombre_producto': ['Producto {}'.format(i) for i in range(1, 21)],
    'Precio': np.random.uniform(10, 100, 20),
    'Cantidad_vendida': np.random.randint(1, 100, 20),
    'Fecha_venta': pd.date_range(start='2023-01-01', periods=20),
    'Edad_estudiante': np.random.randint(18, 25, 20),
    'Calificacion_mate': np.random.randint(60, 100, 20),
    'Calificacion_ciencias': np.random.randint(60, 100, 20),
    'Temperatura': np.random.uniform(0, 30, 20),
    'Humedad': np.random.uniform(20, 90, 20),
})
# Introducir algunos valores NaN aleatoriamente
nan_indices = np.random.choice(df.index, size=5, replace=False)
df.loc[nan_indices, 'Precio'] = np.nan
df.loc[nan_indices, 'Cantidad_vendida'] = np.nan
df.loc[nan_indices, 'Calificacion_mate'] = np.nan

# Imprimir el DataFrame generado
df.head(20)
```

## Identifica los valores faltantes o nulos de cada columna y manéjalos a tu criterio.

```python
df=df.dropna()
#df = df.fillna('No disponible')
df.head(10)
```

## Genera una visualización que muestre las principales medidas estadísticas de cada columna

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient="v")
plt.title('Gráfico de Caja de las Columnas')
plt.xlabel('Columnas')
plt.ylabel('Valor')
plt.show()
```
